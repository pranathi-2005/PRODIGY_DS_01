```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv("/content/EDA -2.csv")
```

```python
print(data)
```

```python

data.info()
```

```python
df_clean = data.dropna(subset=['price'])
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df_clean['item'], df_clean['price'], color='skyblue')
plt.xlabel('Item')
plt.ylabel('Price')
plt.title('Price of Items')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
print(data.columns)
numerical_columns = data.select_dtypes(include=['number']).columns.tolist()
data[numerical_columns].hist(bins=10, figsize=(12, 8), color='skyblue', edgecolor='black')
plt.suptitle('Histograms of Numerical Columns')
plt.show()
```
